```
new_df["closest_bell_tower"] = np.nan
new_df["closest_telus_tower"] = np.nan
new_df["closest_rogers_tower"] = np.nan
```

```
#isp_list = ['Rogers Communications Canada Inc.', 'Bell Mobility Inc.', 'TELUS Communications Inc.','FIDO SOLUTIONS INC.','TELUS Communications Company']

#new_towers_df = towers_df[towers_df[11].isin(isp_list)]
rogers_list = ['Rogers Communications Canada Inc.', 'FIDO SOLUTIONS INC.']
telus_list = ['TELUS Communications Inc.', 'TELUS Communications Company']

rogers_towers = new_towers_df[new_towers_df[11].isin(rogers_list)]
telus_towers = new_towers_df[new_towers_df[11].isin(telus_list)]
bell_towers = new_towers_df[new_towers_df[11] == 'Bell Mobility Inc.']
```

```
bell_towers.head()
```

```
from math import cos, asin, sqrt
def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295     #Pi/180
    a = 0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2
    return 12742 * asin(sqrt(a)) #2*R*asin...
```

```
number_index = 0
for index, row in new_df.iterrows():
  
    #closest rogers
    rogers_distances = []
    for index_2, row_2 in rogers_towers.iterrows():
      d = distance(row['latitude_deg'], row['longitude_deg'], row_2[5], row_2[6])
      rogers_distances.append(d)
    
    new_df.loc[number_index, 'closest_rogers_tower'] = min(rogers_distances)
      
    telus_distances = []
    for index_2, row_2 in telus_towers.iterrows():
      d = distance(row['latitude_deg'], row['longitude_deg'], row_2[5], row_2[6])
      telus_distances.append(d)
      
    new_df.loc[number_index, 'closest_telus_tower'] = None

    bell_distances = []
    for index_2, row_2 in bell_towers.iterrows():
      d = distance(row['latitude_deg'], row['longitude_deg'], row_2[5], row_2[6])
      bell_distances.append(d)
      
    new_df.loc[number_index, 'closest_bell_tower'] = min(bell_distances)
     
    number_index += 1     
```
